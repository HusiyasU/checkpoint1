# Visualisasi Data yang Relevan dengan Buiseness Understanding

Tujuan utama proyek ini adalah untuk mengetahui apa yang menyebabkan popularitas game di Steam. Dengan menganalisis data seperti rating, harga, jumlah ulasan, dan jumlah pemain, proyek ini bertujuan untuk memberikan informasi kepada pengembang dan penerbit game tentang cara meningkatkan daya tarik dan strategi pemasaran game mereka. Selain itu, proyek ini juga bertujuan untuk memperbaiki fitur game dan strategi penetapan harga yang dapat meningkatkan jumlah pemain.

# Distribusi harga game di steam

```python
df["gain"] = pd.to_numeric(df["gain"].str.replace(",", "").str.replace("+", ""), errors='coerce')
df["%gain"] = pd.to_numeric(df["%gain"].str.replace("%", "").str.replace("+", ""), errors='coerce')
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["price"], bins=50, kde=True, color="blue")
plt.xlabel("Harga (USD)")
plt.ylabel("Jumlah Game")
plt.title("Distribusi Harga Game di Steam")
plt.xlim(0, df["price"].max())

plt.show()
```

Grafik di atas menunjukkan distribusi harga game di Steam. Sebagian besar game memiliki harga rendah, dengan puncak distribusi di sekitar 0 hingga 10USD.

Insight awal :     


*   Banyak game yang gratis atau memiliki harga terjangkau.
*   Harga tinggi tampaknya jarang digunakan oleh penerbit game.

*   Bisa jadi game gratis atau murah lebih menarik bagi pemain dan memiliki jumlah pemain lebih banyak.






# Hubungan antara Harga dan jumlah rata-rata pemain

Selanjutnya, analisis hubungan antara harga dan jumlah pemain rata-rata untuk melihat apakah harga memengaruhi popularitas game.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["price"], y=df["avg_players"], alpha=0.5, color="orange")
plt.xlabel("Harga (USD)")
plt.ylabel("Jumlah Pemain Rata-rata")
plt.title("Hubungan antara Harga dan Jumlah Pemain Rata-rata")
plt.ylim(0, df["avg_players"].quantile(0.95))

plt.show()
```

Grafik menunjukkan bahwa sebagian besar game dengan jumlah pemain tinggi memiliki harga rendah atau bahkan gratis.

Insight:


*   Game dengan harga 0 hingga 10USD cenderung memiliki lebih banyak pemain rata-rata.
*   Game dengan harga lebih tinggi memiliki lebih sedikit pemain, meskipun ada beberapa pengecualian.

*   Model freemium atau game murah kemungkinan lebih efektif untuk menarik lebih banyak pemain.






Tindakan yang dapat diambil(termasuk untuk Distribusi harga di steam) :    


*   Pengembang dapat mempertimbangkan strategi harga yang lebih rendah atau model free-to-play dengan monetisasi dalam game.
*   Untuk game berbayar dengan harga tinggi, perlu strategi pemasaran yang lebih kuat atau nilai tambah seperti konten eksklusif.



# Rata rata pemain berdasarkan genre

```python
genre_avg_players = df.groupby("genre")["avg_players"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 18))
sns.barplot(y=genre_avg_players.index, x=genre_avg_players.values, palette="viridis")
plt.xlabel("Rata-rata Pemain")
plt.ylabel("Genre")
plt.title("Rata-rata Pemain Berdasarkan Genre")

plt.show()
```

Grafik menunjukkan bahwa genre tertentu memiliki rata-rata pemain lebih tinggi dibandingkan yang lain.

Insight:


*   Genre dengan jumlah pemain rata-rata tertinggi kemungkinan besar lebih menarik bagi komunitas gamer.
*   Genre seperti Action, Shooter, atau RPG tampaknya lebih populer dibandingkan genre lain seperti Simulation atau Strategy.


*   Pengembang yang ingin menarik lebih banyak pemain bisa mempertimbangkan elemen dari genre populer ini dalam desain game mereka.





Tindakan yang dapat diambil :     


*   Jika targetnya adalah jumlah pemain tertinggi, mengembangkan game dalam genre yang populer bisa menjadi pilihan strategis.
*   ika mengembangkan game di genre yang kurang populer, mungkin perlu strategi pemasaran yang lebih agresif atau fitur unik yang membedakan dari pesaing.



# Game yang kompatibel dengan sistem operasi(Windows, Linux, Mac)

```python
os_counts = {
    "Windows": df["win"].sum(),
    "Mac": df["mac"].sum(),
    "Linux": df["linux"].sum()
}
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=list(os_counts.keys()), y=list(os_counts.values()), palette="pastel")

plt.xlabel("Sistem Operasi")
plt.ylabel("Jumlah Game")
plt.title("Jumlah Game yang Kompatibel dengan Windows, Mac, dan Linux")

plt.show()
```

Insight:


*   Windows memiliki jumlah game kompatibel paling banyak dibandingkan Mac dan Linux.

*   Mac dan Linux memiliki jauh lebih sedikit pilihan game dibandingkan Windows.
*   Pengembang yang ingin menjangkau lebih banyak pemain bisa mempertimbangkan kompatibilitas lintas platform.





Tindakan:


*   Fokus pada Windows Pastikan performa optimal karena mayoritas pemain ada menggunakan os ini.
*   Dukung Mac & Linux Jika Potensial, Gunakan engine lintas platform atau Proton (Steam Play).
*   Porting butuh investasi, analisis permintaan sebelum ekspansi.
